```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Se carga el archivo con los datos de los trayectos de los puntos relevantes.
df = pd.read_csv('Data.csv')
df = df.drop(df.index[1]).reset_index(drop=True).drop(index=0)
df = df[['mass Ax', 'mass Ay', 'mass Bx', 'mass By', 'mass Cx', 'mass Cy', 'mass Dx', 'mass Dy']].astype(float)
df.head()
```

```python
#Ahora vamos a graficar los trayectos de cada punto relevante en gráficos individuales.

plt.style.use("seaborn-v0_8-whitegrid")

fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharex=True, sharey=True)

plots_data = [
    {
        "x": "mass Ax",
        "y": "mass Ay",
        "title": "Trayectoria de Punto A",
        "color": "#1f77b4",
    },
    {
        "x": "mass Bx",
        "y": "mass By",
        "title": "Trayectoria de Punto B",
        "color": "#ff7f0e",
    },
    {
        "x": "mass Cx",
        "y": "mass Cy",
        "title": "Trayectoria de Punto C",
        "color": "#2ca02c",
    },
    {
        "x": "mass Dx",
        "y": "mass Dy",
        "title": "Trayectoria de Punto D",
        "color": "#d62728",
    },
]

for i, data in enumerate(plots_data):
    ax = axs.flat[i]

    ax.plot(
        df[data["x"]],
        df[data["y"]],
        marker="o",
        markersize=5,
        linestyle="-",
        linewidth=2,
        color=data["color"],
        label=data["title"].split()[-1],  
    )

    ax.set_title(data["title"], fontsize=13, fontweight="bold", pad=10)
    ax.grid(True, linestyle="--", alpha=0.6)

    if i >= 2:
        ax.set_xlabel("Eje X", fontsize=11)
    if i % 2 == 0:
        ax.set_ylabel("Eje Y", fontsize=11)

fig.suptitle(
    "Comparación de Trayectorias (Puntos A, B, C, D)",
    fontsize=16,
    fontweight="bold",
    y=0.98,
)

plt.tight_layout()
plt.show()
```
